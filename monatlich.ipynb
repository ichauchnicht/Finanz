```python
inputpath = r"D:\privat_neu\finanz\ausgaben\Exoprte\Toshl export Dec 1, 2022 - Dec 31, 2023.csv"
outputpathxls = r"D:\temp\test.xls"
outputfolder = r"D:\temp"
```

```python
import os
import pandas as pd
import matplotlib.pyplot as plt
pd.set_option('float_format', '{:.2f}'.format)



data = pd.read_csv(inputpath)
data.Date =pd.to_datetime(data.Date)
#data['ym']=data.Date.dt.year+'_'+ data.Date.dt.month
data = data.rename(columns={'Income amount':'Income', 'Expense amount':'Expense', 'In main currency':'In_main_currency'})
data.Income = pd.to_numeric(data.Income.str.replace(',',''))
data.Expense = pd.to_numeric(data.Expense.str.replace(',',''))
data.In_main_currency = pd.to_numeric(data.In_main_currency.str.replace(',',''))


#put all expences in main currency and negative
data['allExpencesInSEK'] = 0
data.loc[data.Expense > 0, 'allExpencesInSEK'] = data.In_main_currency
data['allExpencesInSEK'] = data['allExpencesInSEK'] *-1

data['Year_Month'] = data.Date.dt.to_period('M')
data
#In main currency In_main_currency
```

```python
numberOfMonth = data['Year_Month'].nunique()
print(numberOfMonth)
```

```python

pivot = pd.pivot_table(
    data,
    index=["Category", "Tags"],
    values=["allExpencesInSEK"],
    aggfunc='sum',
    fill_value=0,
    columns=["Year_Month"],
    margins=True)
pivot.columns = pivot.columns.get_level_values(1)
pivot['avg'] = pivot['All'] / numberOfMonth
out = os.path.join(outputfolder, 'allmonth.csv')
pivot.to_csv(out)

pivot
```

```python
pivotincome = pd.pivot_table(
    data,
    index=["Category"],
    values=["Income"],
    aggfunc='sum',
    fill_value=0,
    columns=["Year_Month"],
    margins=True)

pivotincome.columns = pivotincome.columns.get_level_values(1)
pivotincome['avg'] = pivotincome['All'] / numberOfMonth
allIncomes = pivotincome.loc[['Salary', 'Grants' , 'Reimbursements'] , : ].sum()
out = os.path.join(outputfolder, 'allmonth_withincome.csv')

pivotincome.to_csv(out)
```

```python
pivotincome = pd.pivot_table(
    data,
    index=["Category"],
    values=["Income"],
    aggfunc='sum',
    fill_value=0,
    columns=["Year_Month"],
    margins=True)
pivotincome.columns = pivotincome.columns.get_level_values(1)
pivotincome['avg'] = pivotincome['All'] / numberOfMonth
pivotincome
```

```python
pivotGeneral = pd.pivot_table(
    data,
    index=["Category"],
    values=["allExpencesInSEK"],
    aggfunc='sum',
    fill_value=0,
    columns=["Year_Month"],
    margins=True)


pivotGeneral.columns = pivotGeneral.columns.get_level_values(1)
pivotGeneral['avg'] = pivotGeneral['All'] / numberOfMonth

allIncomes = pivotincome.loc[['Salary', 'Grants' , 'Reimbursements'] , : ].sum()

allIncomes.name = 'all incomes'
allExpences = pivotGeneral.loc['All']
over = allIncomes + allExpences #expenses are negative!
over.name = 'left'

sparquote = pivotGeneral.loc['Sparen'] / allIncomes
sparquote.name = 'Sparquote'

reineAusgaben = allExpences - pivotGeneral.loc['Sparen']

reineAusgaben.name = 'reine Ausgaben'

pivotGeneral= pivotGeneral.append(allIncomes, ignore_index =False, sort= False)
pivotGeneral= pivotGeneral.append(over, ignore_index =False, sort= False)
pivotGeneral= pivotGeneral.append(sparquote, ignore_index =False, sort= False)
pivotGeneral= pivotGeneral.append(reineAusgaben, ignore_index =False, sort= False)
      

out = os.path.join(outputfolder, 'allmonth_general.csv')
pivotGeneral.to_csv(out)

pivotGeneral
```
